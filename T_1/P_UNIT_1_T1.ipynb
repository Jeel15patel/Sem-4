# FIND OUTLIRES

```python
# def fund_outline(ds,col)

import pandas as pd
auto = pd.read_csv('auto-mpg.csv')
auto.describe()
```

```python
def find_outliers(ds,col):
    qurd_1 = ds[col].quantile(0.25)
    qurd_3 = ds[col].quantile(0.75)
    IQR = qurd_3-qurd_1
    low_val = qurd_1 - 1.5 * IQR
    high_val = qurd_3 + 1.5 * IQR
    ds = ds.loc[(ds[col]<low_val)|(ds[col]>high_val)]
    print("Lower_Value",low_val)
    print("High_Value",high_val)
    print("Qurd_1",qurd_1)
    print("Qurd_3",qurd_3)
    print("IQR",IQR)
    return ds
find_outliers(auto,"acceleration")  
```

# duplicated()
Duplicated gives duplicated ROW. Cond. is Both Row have same data

```python
import pandas as pd

data = {
    "A" : ["Team A" , "Team B" , "Team B" , "Team C" , "Team A"],
    "B" : [50 , 40 , 40 , 30 , 50],
    "C" : [True , False , False , False , True]
}

df = pd.DataFrame(data)
print(df)

print("----------")

dudf = df.duplicated()
print(dudf)
```

# drop_duplicates()
Remove duplicates data from file.

```python
import pandas as pd

data = {
    "A" : ["Team A" , "Team B" , "Team B" , "Team C" , "Team A"],
    "B" : [50 , 40 , 40 , 30 , 50],
    "C" : [True , False , False , False , True]
}

df = pd.DataFrame(data)
print(df)

print("----------")

dudf = df.drop_duplicates()
print(dudf)

print("----------")

redf = dudf.reset_index()
print(redf)
```
